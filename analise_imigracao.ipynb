```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/bibliotecas_visualizacao/refs/heads/main/Dados/imigrantes_canada.csv'
dados_imigracao = pd.read_csv(url)
dados_imigracao.set_index('País', inplace= True)
dados_imigracao.head()
```

```python
dados_imigracao.info()
```

```python
anos = list(map(str,(range(1980,2014))))
anos
```

```python
dados_brasil = dados_imigracao.loc['Brasil',anos]
dados_brasil = pd.DataFrame(dados_brasil,dtype= int).reset_index().rename(columns={'index':'ano','Brasil':'imigrantes'})
dados_brasil
```

```python
fig , ax = plt.subplots(figsize = (8,6))
ax.set_title('Evolução da imigração de brasileiro para o Canadá\n1980 a 2013', loc='left', fontsize=15 )
ax.plot(dados_brasil.ano,dados_brasil.imigrantes, lw = 3, color = 'g')
ax.set_xlabel('Fonte: curso Alura',  fontsize = 10, loc = 'right')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.spines[['top', 'right']].set_visible(False)
ax.grid(True, color ='grey', ls = '--',lw = 0.5, alpha = 0.5)
imagem = ax.get_figure()
imagem.savefig('imigracao_brasil.png',transparent=False, dpi = 300, bbox_inches= 'tight')
plt.show()
```

```python
fig = plt.figure(figsize = (10,6))

ax1 = plt.subplot2grid((1,3), (0,0), colspan=2)
ax2 = plt.subplot2grid((1,3), (0,2), colspan=1)

ax1.plot(dados_brasil.ano,dados_brasil.imigrantes, lw = 3, color = 'g')
ax1.xaxis.set_major_locator(plt.MultipleLocator(5))
ax1.spines[['top', 'right']].set_visible(False)
ax1.grid(True, color ='grey', ls = '--',lw = 0.5, alpha = 0.5)

ax2.boxplot(dados_brasil.imigrantes, patch_artist=True, boxprops={"facecolor": "green", "alpha":0.5},medianprops={"color": "black", "linewidth": 0.5}, widths=0.5 )
ax2.spines[['top','right']].set_visible(False)
ax2.set_xlabel('Fonte: curso Alura',  fontsize = 10, loc = 'right')
ax2.set_xticks([])



fig.suptitle('Evolução da imigração de brasileiro para o Canadá 1980 a 2013',  fontsize=15, x=0.35)
plt.show()
```

```python
paises = dados_america_sul.drop(['Continente','Região','Total'], axis = 1).T
cores = ['black','y','r','g']

fig , axs = plt.subplots(2,2,figsize = (10,10))

for ax,pais,cor in zip(axs.flat, paises.head(4),cores):
  ax.plot(paises.index,paises[pais], lw = 3, color = cor)
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.spines[['top', 'right']].set_visible(False)
  ax.grid(True, color ='grey', ls = '--',lw = 0.5, alpha = 0.5)
  ax.set_title(pais)
  ax.set_ylim(0,7000)
axs[1,1].set_xlabel('Fonte: curso Alura',  fontsize = 10, loc = 'right')
fig.suptitle('Maiores imigrações para o Canadá da América do Sul\n1980 a 2013', ha='left', fontsize=15, x =0.12)
plt.show()
```

```python
cores = []

for pais in dados_america_sul.sort_values(by='Total',ascending=True).Total.index:
  if pais == 'Brasil':
    cores.append('green')
  else:
    cores.append('silver')


fig , ax = plt.subplots(figsize = (10,8))

ax.barh(dados_america_sul.sort_values(by='Total',ascending=True).Total.index, dados_america_sul.sort_values(by='Total',ascending=True).Total, color = cores)

for i,v in enumerate(dados_america_sul.sort_values(by='Total',ascending=True).Total):
  ax.text(v + 30, i, v, fontsize = 10, ha= 'left', va = 'center')

ax.spines[:].set_visible(False)
ax.set_xticks([])
ax.tick_params(axis ='both', length = 0 )
ax.set_xlabel('Fonte: curso Alura',  fontsize = 10, loc = 'right')
fig.suptitle('Brasil é o a 4ª maior imigração para o Canadá\n1980 a 2013', ha='left', fontsize=15, x =0.12, y= 0.91)
plt.show()
```

```python
# Criando uma figura
fig = go.Figure()

# Adicionando a linha do gráfico e definindo a espessura da linha
fig.add_trace(
    go.Scatter(x=[dados_brasil['ano'].iloc[0]], y=[dados_brasil['imigrantes'].iloc[0]], mode='lines', name='Imigrantes', line=dict(width=4))
)

# Definindo as configurações de layout
fig.update_layout(
    title=dict(
        text='<b>Imigração do Brasil para o Canadá no período de 1980 a 2013</b>',
        x=0.2,
        xanchor='left',
        font=dict(size=20)),
    xaxis=dict(range=[1980, 2013], autorange=False),
    yaxis=dict(range=[0, 3000], autorange=False),
    width=1000,
    height=500,

    updatemenus=[{
                'type':'buttons',
                'showactive':False,
                "direction": "left",
                "pad": {"r": 40, "t": 350},
                'buttons':[{
                          'label':'Play',
                          'method':'animate',
                          'args':[None, {'frame': {'duration': 100, 'redraw': False}, 'fromcurrent': True}]
                          },
                        {
                          'label':'Pause',
                          'method':'animate',
                          'args':[[None], {'frame': {'duration': 0, 'redraw': False}, 'mode':'immediate',"transition": {"duration": 0}}]
                         }]
              }])

# Definir as configurações de animação
frames = [go.Frame(data=[go.Scatter(x=dados_brasil['ano'].iloc[:i+1], y=dados_brasil['imigrantes'].iloc[:i+1])]) for i in range(len(dados_brasil))]
fig.frames = frames

# Mostrando a figura
fig.show()
```

```python
dados_imigracao_v2 = pd.DataFrame(columns =['ano',	'imigrantes',	'pais',	'Continente',	'Região'])

for pais in dados_imigracao.index:

  dados = dados_imigracao.loc[pais,anos]
  dados = pd.DataFrame(dados,dtype= int).reset_index().rename(columns={'index':'ano',pais:'imigrantes'})
  dados.ano = dados.ano.astype(int)
  dados['pais'] = pais
  dados = dados.join(dados_imigracao[['Continente','Região']], on= 'pais' )
  dados_imigracao_v2 = pd.concat([dados_imigracao_v2,dados])

dados_imigracao_v2.ano = dados_imigracao_v2.ano.astype(dtype = int)
dados_imigracao_v2.imigrantes = dados_imigracao_v2.imigrantes.astype(dtype = int)
dados_imigracao_v2
```

```python
#Paises com imigrações maiores que 10000
print(f'Paises: ',dados_imigracao_v2.query('imigrantes > 10000').pais.unique())
print(f'\nNúmeros de ocorrencias maiores de 10000:', len(dados_imigracao_v2.query('imigrantes > 10000').pais))
```

```python
# Criando uma figura
fig = go.Figure()

# Adicionando a linha do gráfico e definindo a espessura da linha
for p in dados_imigracao_v2.pais.unique():
  fig.add_trace(
      go.Scatter(x= [dados_imigracao_v2.query('pais == @p').ano.iloc[0]],
                 y= [dados_imigracao_v2.query('pais == @p').imigrantes.iloc[0]],
                 name = p,
                 mode='lines',
                 visible=True,
                 ))

# Criando o botão por região
botao = []
for regiao in dados_imigracao_v2.Região.sort_values().unique():
    traces = [True if dados_imigracao_v2.query('pais == @pais')['Região'].iloc[0] == regiao else False
                      for pais in dados_imigracao_v2.pais.unique()]

    botao.append(dict(label = regiao,
                        method='update',
                        args=[{'visible': traces},
                              {'title': f'Imigração para o Canadá: Região {regiao}'}]))

# Posicionando os botões de play e pause
fig.update_layout(    title=dict(
                      text='<b>Imigração para o Canadá no período de 1980 a 2013</b>',
                      x=0.2,
                      xanchor='left',
                      font=dict(size=20)),
                      xaxis=dict(range=[1980, 2013], autorange=False),
                      yaxis=dict(range=[0, 10000], autorange=False),
                      width=1200,
                      height=700,

                  updatemenus =  [{
                                  'type' : "dropdown",
                                  'direction' : "down",
                                  'buttons' : botao,
                                  'x' : 1.09,
                                  'xanchor' : "left",
                                  'y' : 1.113,
                                  'yanchor' : "top",
                                  },
                                  {
                                  'type':'buttons',
                                  'showactive':False,
                                  "direction": "left",
                                  "pad": {"r": 50, "t": 550},
                                  'buttons':[
                                            {
                                            'label':'Play',
                                            'method':'animate',
                                            'args':[None, {'frame': {'duration': 100, 'redraw': False}, 'fromcurrent': True}]
                                            },
                                            {
                                            'label':'Pause',
                                            'method':'animate',
                                            'args':[[None], {'frame': {'duration': 0, 'redraw': False}, 'mode':'immediate',"transition": {"duration": 0}}]
                                            }
                                            ]
                                    }]
                  )

# Definindo as configurações de animação
frames = []

for i in range(len(dados_imigracao_v2.query('pais == "Brasil"'))):
  frame_data = []
  for p in dados_imigracao_v2.pais.unique():
           frame_data.append(go.Scatter(x = dados_imigracao_v2.query('pais == @p').ano.iloc[:i+1],
                                    y = dados_imigracao_v2.query('pais == @p').imigrantes.iloc[:i+1]
                                    ))
  frame = go.Frame(data=frame_data)
  frames.append(frame)
fig.frames = frames

# Mostrando a figura
fig.show()
```
